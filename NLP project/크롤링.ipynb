```python
# 잡코리아 사이트에서 데이터/IT 분야의 자소서 중 전문가 평점이 있는 자소서들만 필터링 > 342개 > url만 크롤링
from selenium import webdriver
from selenium.webdriver.common.by import By
import time 

driver = webdriver.Chrome()

url_list = ['https://www.jobkorea.co.kr/starter/PassAssay?FavorCo_Stat=0&Pass_An_Stat=1&OrderBy=2&EduType=0&WorkType=0&schPart=10031&isSaved=1&Page=' + str(x) for x in range(1, 18)]
links = []
for url in url_list:
    driver.get(url)
    time.sleep(3)
    question_items = driver.find_elements(By.CSS_SELECTOR, 'div.item.question')
    for item in question_items:
        a_tag = item.find_element(By.TAG_NAME, 'a')
        href = a_tag.get_attribute('href')
        links.append(href)

driver.quit()
```

```python
from selenium import webdriver
import bs4
import csv

# 크롬 드라이버 설정
driver = webdriver.Chrome()

# 자소서 url 리스트
file_path = 'urls.csv'
with open(file_path, "r", encoding="utf-8-sig") as f:
    url_reader = csv.reader(f)
    urls = [row[0] for row in url_reader][160:240]


def crawling_url(url):
    driver.get(url)
    response = driver.page_source
    one_page = bs4.BeautifulSoup(response, 'html.parser')
    
    # 불필요 요소 제거
    for p_element in one_page.find_all('p', {'class': 'txSpllChk'}):
        p_element.extract()

    for span_element in one_page.find_all('span', {'class': 'sup'}):
        span_element.extract()

    for a_element in one_page.find_all('a'):
        a_element.extract()
        
    # 공백 제거 및 텍스트만 추출
    kwd_list = []
    for text in one_page.find_all('div', {'class': 'tx'}):
        kwd_list.append(text.get_text(separator=' ', strip=True))
    
    return kwd_list

all_kwd_lists = []
for kwd_url in urls:
    kwd_list = crawling_url(kwd_url)
    all_kwd_lists.append(kwd_list)

driver.quit()
```

```python
# 크롤링한 데이터를 CSV 파일로 저장
output_file_path = 'crawled_data0.csv'

with open(output_file_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    for i, kwd_list in enumerate(all_kwd_lists):
        writer.writerow(kwd_list)

print(f" {output_file_path} 저장완료 ")
```

```python
# join
join_list = []
input_file_path = 'crawled_data0.csv'
with open(input_file_path, 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    
    for row in reader:
        sen = ' '.join(row)
        join_list.append(sen)

# 새 파일로 저장        
output_file_path = 'crawled_data1.csv'
with open(output_file_path, 'w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    
    for sen in join_list:
        writer.writerow([sen])
        
print(f"{output_file_path} 저장 완료")
```

```python
import os
import pandas as pd

file_path = r'C:/Users/user/Desktop/hansol/3rd_project/new/NLP-project/NLP project/crawled_data0.csv'

if os.path.exists(file_path):
    # 인코딩을 지정하여 CSV 파일을 읽음
    try:
        df = pd.read_csv(file_path, header=None, encoding='utf-8', on_bad_lines='skip')
    except UnicodeDecodeError:
        df = pd.read_csv(file_path, header=None, encoding='cp949', on_bad_lines='skip')
    
    total_length = len(df)
    print(f'CSV 파일의 총 행 수: {total_length}')
    
    # 각 행의 길이 계산
    line_lengths = df.apply(lambda row: len(','.join(row.astype(str))), axis=1)

    # 각 행의 길이 출력
    length_list = []
    for index, length in line_lengths.items():
        print(f'{index + 1}번째 행의 길이: {length}')
        length_list.append(length)
    
    # 길이가 2990을 초과하는 행의 개수 계산
    count_over_2990 = (line_lengths > 2990).sum()
    print(f'길이가 2990을 초과하는 행의 개수: {count_over_2990}')

else:
    print(f'파일이 존재하지 않습니다: {file_path}')
```
