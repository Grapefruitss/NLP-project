```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import UnexpectedAlertPresentException
import time
import csv
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome( options=options)
```

```python
jobko = ["https://www.jobkorea.co.kr/starter/PassAssay?FavorCo_Stat=0&Pass_An_Stat=1&OrderBy=2&EduType=0&WorkType=0&schPart=10031&isSaved=0&Page=7"]

for url in jobko:
    try:
        driver.get(url)
        count = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '#container > div.stContainer > div.starListsWrap.ctTarget > ul > li')))
    except Exception as e:
        print(f'{e}')
driver.quit()
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=options)

assay_list = []


script = ''' var element = document.querySelectorAll('dl > dd');  element.forEach((el)=>{el.classList.add("show")})    '''
for url in url_list:

    try:
        driver.get(url)

        WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.TAG_NAME, 'body')))
        driver.execute_script(script)


        sup_elements = driver.find_elements(By.CSS_SELECTOR, '.sup')
        for elem in sup_elements:
            driver.execute_script('arguments[0].remove()', elem)


        txSpllChk_elements = driver.find_elements(By.CLASS_NAME, 'txSpllChk')
        for elem in txSpllChk_elements:
            driver.execute_script('arguments[0].remove()', elem)


        advice_elements = driver.find_elements(By.CSS_SELECTOR, '.advice')
        for elem in advice_elements:
            driver.execute_script('arguments[0].remove()', elem)


        dltag_text = driver.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/dl').text
        assay_list.append(dltag_text)

    except Exception as e:
        print(f'{e}')





driver.quit()


```

```python
import re

# 크롬 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# 드라이버 초기화 및 페이지 로드
driver = webdriver.Chrome(options=options)
driver.get('https://papago.naver.com/?sk=ko&tk=en')

# 번역 결과를 저장할 리스트 초기화
arr = []

# assay_list는 크롤링해온 자소서의 리스트입니다.

for assay in assay_list:
    try:
        # 입력 필드 찾기
        input_field = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="txtSource"]'))
        )
        input_field.clear()

        # 문장 분할
        sentences = re.split(r'(?<=[.!?])\s+', assay)
        translated_sentences = []

        for sentence in sentences:
            input_field.send_keys(sentence)

            # 번역된 텍스트 대기
            output_xpath = '//*[@id="txtTarget"]/span'
            WebDriverWait(driver, 60).until(
                EC.presence_of_element_located((By.XPATH, output_xpath))
            )
            output = driver.find_element(By.XPATH, output_xpath)
            translated_text = output.text.strip()
            translated_sentences.append(translated_text)

            input_field.clear()
            time.sleep(1)  # 서버 부담을 줄이기 위해 대기 시간 추가

        # 번역된 문장을 하나로 합치기
        translated_text = ' '.join(translated_sentences)
        arr.append(translated_text)

    except Exception as e:
        print(f'Error occurred: {e}')

# 드라이버 종료
driver.quit()

# 결과를 파일에 저장
with open('en_assay3.txt', 'w', encoding='utf-8') as f:
    for txt in arr:
        f.write(txt + '\n')
```
