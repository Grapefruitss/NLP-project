```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome( options=options)
```

```python
jobko = ["https://www.jobkorea.co.kr/starter/PassAssay?FavorCo_Stat=0&Pass_An_Stat=1&OrderBy=0&EduType=0&WorkType=0&schPart=10031&isSaved=0&Page=17"]
for i in jobko:
    try:
        driver.get(i)
        count = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '#container > div.stContainer > div.starListsWrap.ctTarget > ul > li')))
    except Exception as e:
        print(f'{e}')
```

```python
len(count)
```

```python
url_list = []
for j in range(len(count)):
    driver.get(i)
    wait = WebDriverWait(driver, 30)
    driver.execute_script(f"document.querySelectorAll('.logo')[{j}].click();")
    url_list.append(driver.current_url)
driver.quit()

print(url_list)
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=options)

assay_list = []


script = ''' var element = document.querySelectorAll('dl > dd');  element.forEach((el)=>{el.classList.add("show")})    '''
for url in url_list:

    try:
        driver.get(url)
        time.sleep(15)
        driver.execute_script(script)


        sup_elements = driver.find_elements(By.CSS_SELECTOR, '.sup')
        for elem in sup_elements:
            driver.execute_script('arguments[0].remove()', elem)


        txSpllChk_elements = driver.find_elements(By.CLASS_NAME, 'txSpllChk')
        for elem in txSpllChk_elements:
            driver.execute_script('arguments[0].remove()', elem)


        advice_elements = driver.find_elements(By.CSS_SELECTOR, '.advice')
        for elem in advice_elements:
            driver.execute_script('arguments[0].remove()', elem)


        dltag_text = driver.find_element(By.XPATH, '//*[@id="container"]/div[2]/div[3]/dl').text
        assay_list.append(dltag_text)

    except Exception as e:
        print(f'{e}')

driver.quit()
```

```python
assay_list = [i.replace('\n', ' ') for i in assay_list]
assay_list
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=options)


arr = []

for i, assay in enumerate(assay_list):
    driver.get('https://papago.naver.com')
    try:
        input = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="txtSource"]')))
        input.clear()
        input.click()
        input.send_keys(assay)
        time.sleep(40)
        spans = driver.find_elements(By.XPATH, '//*[@id="txtTarget"]/span')


        for j in spans:
            arr.append(j.text)
    except Exception as e:
        print(f'{e}')

driver.quit()
```

```python
with open('data.txt', 'w') as f:
    for i in arr:
        f.write(i + '\n')
```
